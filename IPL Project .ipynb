```python
#Top 5 batsmen based on past 3 years boundary % (fours and sixes).

#Grouping the data by batsman_name and then summing it.
group_name = fact_batting.groupby('batsmanName').sum('6s')
#Adding the 4s and 6s to get the total boundaries by each batsman.
group_name['totalBoundaries']=group_name['4s']+group_name['6s']
#boundary%
group_name['boundary%'] = (group_name['totalBoundaries']/group_name['balls'])*100
#Sorting the values by the number of boundaries and also,
#defining a condition that the batsman must have faced 150 balls total.
sorted_data = group_name[group_name['balls']>=150].sort_values(by= ['boundary%'],ascending = False)
#fetching only the columns we need.
sorted_data.head()[['balls','totalBoundaries','boundary%']]
```

```python
#Top 5 bowlers based on past 3 years dot ball %.
#grouping the data as per baller names:
fact_balling['balls'] = fact_balling['overs']*6
#Summing the results to get the number of dot balls by each baller:
grouped_data = fact_balling.groupby('bowlerName').sum('0s')
#finding the dot ball %
grouped_data['dot_balls_percentage'] = (grouped_data['0s']/grouped_data['balls'])*100
#Defining a condition that the baller should have atleast balled 180 balls:
final_group = grouped_data[grouped_data['balls']>=180]
final_group['dot_balls_percentage'].nlargest(5)
```

```python
#Top 4 teams based on past 3 years winning %.
#Num of matches won grouped by team:
matches_won = dim_match_summary.groupby('winner').size()
#Let us find the total number of matches played:
team1_played = dim_match_summary.groupby('team1').count()['match_id']
team2_played = dim_match_summary.groupby('team2').count()['match_id']
total_matches = team1_played+team2_played
#Calculating the winning %:
((matches_won/total_matches)*100).nlargest(4)
```

```python
#Top 2 teams with the highest number of wins achieved by chasing targets over the past 3 years.
chase_win_teams = dim_match_summary[dim_match_summary['team2'] == dim_match_summary['winner']]
chase_win_teams.groupby('winner').size().nlargest(3)
```

```python
#Top 10 batsmen based on past 3 years batting average. (min 60 balls faced in each season)

# First we create a new df with matchDate merged into fact_batting:
bat_with_date = fact_batting.merge(dim_match_summary[['match_id','matchDate']], how = 'inner', on = 'match_id')
# Let's add the season column too:
bat_with_date['Season'] = pd.to_datetime(bat_with_date['matchDate']).dt.year
# Let's get the total runs by each player:
batsman_60 = bat_with_date.groupby(['Season','batsmanName']).filter(lambda x:(x['balls'].sum()>=60).all())
# Here we are specifying the condition that we only need those players who played in all the last three seasosn:
seasons_played = batsman_60.groupby('batsmanName')['Season'].nunique()
batsman_with_3_season = seasons_played[seasons_played == 3].index
new_batsman_60 = batsman_60[batsman_60['batsmanName'].isin(batsman_with_3_season)]
# Now calculate total runs and outs for these players
total_runs = new_batsman_60.groupby('batsmanName')['runs'].sum()
total_outs = new_batsman_60[new_batsman_60['out/not_out'] == 'out'].groupby('batsmanName').size()
# Here we concatenated the two df's:
batting_stats = pd.concat([total_runs,total_outs],axis = 1)
# Renaming the columns for better understanding:
batting_stats.columns = ['Total_runs',"Total_out"]
# Last we can find the batting average:
batting_stats['batting_average'] = round(batting_stats['Total_runs']/batting_stats['Total_out'],2)
# now we can see the top 10:
batting_stats['batting_average'].nlargest(10)
```

```python
#Top 10 batsmen based on past 3 years strike rate (min 60 balls faced in each season)
#Creating a new df by using the old df we created in the above questions.
#We are here the grouping them by season and bastsman then finding the sum of balls for each pair.
nameball = new_batsman_60.groupby(['Season','batsmanName'])['balls'].sum().reset_index()
#We are here the grouping them by season and bastsman then finding the mean or avg of SR for each pair.
namesr = new_batsman_60.groupby(['Season','batsmanName'])['SR'].mean().reset_index()
#Concatinating the above two df 
new_bat = pd.concat([nameball,namesr],axis = 0)
# lastly we can use the new_bat df to group only with names and then finding the top 10:
new_bat.groupby('batsmanName')['SR'].mean().nlargest(10)
```

```python
#Top 10 bowlers based on past 3 years bowling average. (min 60 balls bowled in each season)
#Merge or join the date column into the fact_balling table:
fact_balling_date = fact_balling.merge(dim_match_summary[['match_id','matchDate']],how = 'inner', on = 'match_id')
#Extract Year from the matchDate column:
fact_balling_date['Season'] = pd.to_datetime(fact_balling_date['matchDate']).dt.year
#Now we need to filter only those players who have balled over 60 balls every season
new_fact_balling_date = fact_balling_date.groupby(['Season','bowlerName']).filter(lambda x: (x['overs'].sum()>=10).all())
#Specify the condition that the player should have played all the seasons:
season_played  = new_fact_balling_date.groupby('bowlerName')['Season'].nunique()
season_played_3 = season_played[season_played == 3].index
#Getting only those players who have played all the seaons:
new_fact_balling = fact_balling_date[fact_balling_date['bowlerName'].isin(season_played_3)]
#Now we need to calculate the bowling average:
total_runs_wickets = new_fact_balling.groupby('bowlerName')[['runs','wickets']].sum('runs')
bowling_average = round((total_runs_wickets['runs']/total_runs_wickets['wickets']),2)
bowling_average.nsmallest(10)
```

```python
#Top 10 bowlers based on past 3 years economy rate. (min 60 balls bowled in each season)
#mergeing the tables to get the season:
new_fact_balling = fact_balling.merge(dim_match_summary[['match_id','matchDate']],how = 'inner', on = 'match_id')
#Extracting the year(seaosn) from the date column:
new_fact_balling['season'] = pd.to_datetime(new_fact_balling['matchDate']).dt.year
#Let's define the condition that the players should have balled 60> more balls:
new_fact_balling_60 = new_fact_balling.groupby(['season','bowlerName']).filter(lambda x:(x['overs'].sum()>=10))
#Let's start the process of filtering out only those players who have playes all three season:
season_played = new_fact_balling_60.groupby('bowlerName')['season'].nunique()
season_played_3 = season_played[season_played==3].index
new_df = new_fact_balling_60[new_fact_balling_60['bowlerName'].isin(season_played_3)]
#Now all we need is to group by the names and start finding the economy rates:
final_df = new_df.groupby('bowlerName').agg({'overs': 'sum', 'runs': 'sum'})
final_df['economy'] = final_df['runs'] / final_df['overs']
final_df['economy'].nsmallest(10)
```
